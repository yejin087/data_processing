```python
#경상남도 일반음식점 csv파일 dataframe으로 읽어들이기
import pandas as pd


file = pd.read_csv("경상남도_일반음식점 현황.csv",encoding='cp949')
restaurant_types = file['업태'].value_counts()
print(restaurant_types)
```

```python
import platform
import matplotlib
from matplotlib import font_manager, rc

# '-' 기호 보이게 하기
matplotlib.rcParams['axes.unicode_minus'] = False

# 운영 체제마다 한글이 보이게 하는 설정
# 윈도우
if platform.system() == 'Windows':
    path = "c:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
restaurant_types.plot(kind='barh', y="Sales", x="Name")
plt.show()
```

```python
#한식 행 삭제

restaurant_types = restaurant_types.drop(labels=['한식'])
restaurant_types.plot(kind='barh', y="Sales", x="Name")
plt.show()
```
